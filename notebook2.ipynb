```python
# Notebook 2 - DiVincenzo Criteria Lab
# IBM Qiskit Fall Fest 2025

# Step 1: Install and import Qiskit
!pip install qiskit --quiet
!pip install qiskit-aer --quiet
!pip install pylatexenc --quiet

from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt

# Step 2: Criterion 1 - Scalable system with well-characterized qubits
print("### Criterion 1: Scalable system with well-characterized qubits ###")
qc1 = QuantumCircuit(2)
qc1.h(0)     # Hadamard on qubit 0
qc1.cx(0, 1) # CNOT between qubit 0 and 1
display(qc1.draw('mpl'))

# Step 3: Criterion 2 - Ability to initialize qubits
print("\n### Criterion 2: Ability to initialize qubits ###")
qc2 = QuantumCircuit(2)
qc2.reset([0, 1])  # Reset both qubits to |0>
qc2.x(1)            # Flip qubit 1 to |1>
display(qc2.draw('mpl'))

# Step 4: Criterion 3 - Long decoherence times (simulate statevector)
print("\n### Criterion 3: Long decoherence times relative to gate operation ###")
sv = Statevector.from_instruction(qc2)
print("Statevector:", sv)
plot_bloch_multivector(sv)

# Step 5: Criterion 4 - Universal set of quantum gates
print("\n### Criterion 4: Universal set of quantum gates ###")
qc3 = QuantumCircuit(2)
qc3.h(0)
qc3.t(0)
qc3.cx(0, 1)
display(qc3.draw('mpl'))

# Step 6: Criterion 5 - Qubit-specific measurement capability
print("\n### Criterion 5: Measurement of qubits ###")
qc4 = QuantumCircuit(2)
qc4.h(0)
qc4.cx(0,1)
qc4.measure_all()
backend = Aer.get_backend('qasm_simulator')
result = backend.run(qc4, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)

# Step 7: Summary
print("\n### Summary ###")
print("""
In this notebook, we demonstrated the five DiVincenzo criteria for building a quantum computer:
1. Scalable system with well-characterized qubits
2. Ability to initialize qubits
3. Long decoherence times relative to gate operation
4. Universal set of quantum gates
5. Qubit-specific measurement capability

We implemented small Qiskit circuits for each criterion, visualized the states, and performed measurements.
""")
```
