```python

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Initialize single qubit in |0> state
qc = QuantumCircuit(1)
state = Statevector.from_instruction(qc)

# Plot Bloch sphere
plot_bloch_multivector(state)
```

```python
# X rotation
qc_x = QuantumCircuit(1)
qc_x.rx(np.pi/2, 0)
state_x = Statevector.from_instruction(qc_x)
plot_bloch_multivector(state_x)

# Y rotation
qc_y = QuantumCircuit(1)
qc_y.ry(np.pi/2, 0)
state_y = Statevector.from_instruction(qc_y)
plot_bloch_multivector(state_y)

# Z rotation
qc_z = QuantumCircuit(1)
qc_z.rz(np.pi/2, 0)
state_z = Statevector.from_instruction(qc_z)
plot_bloch_multivector(state_z)
```

```python
qc_h = QuantumCircuit(1)
qc_h.h(0)
state_h = Statevector.from_instruction(qc_h)
plot_bloch_multivector(state_h)
```

```python
# Simulate measurement using probabilities
state = Statevector.from_instruction(qc_h)
probs = state.probabilities_dict()
print("Measurement probabilities:", probs)

# Histogram plot
plot_histogram(probs)
```

```python
qc_arbitrary = QuantumCircuit(1)
qc_arbitrary.rx(np.pi/4, 0)
qc_arbitrary.ry(np.pi/3, 0)
state_arbitrary = Statevector.from_instruction(qc_arbitrary)
plot_bloch_multivector(state_arbitrary)
```
